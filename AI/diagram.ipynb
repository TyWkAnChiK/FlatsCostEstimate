```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла
data = pd.read_csv('dataSUPER (1).csv', low_memory=False, lineterminator='\n').drop(['id', 'Apartment type', 'Region'], axis=1)
```

```python
data.keys()
```

```python
def BuiltDiagram(name):
    name=str(name)
    # Подсчет количества одинаковых строк в первом столбце
    counts = data[name].value_counts()
    plt.figure(figsize=(20, 6)) 
    # Построение гистограммы
    plt.bar(counts.index, counts.values)


    # Настройка осей и заголовка
    plt.xlabel(name)
    plt.ylabel('Количество')
    plt.title('Гистограмма')

    # Поворот меток по оси X для лучшей читаемости
    plt.xticks(rotation=90)

    # Отображение гистограммы
    plt.show()
    for brand, count in counts.items():
        print(f"{brand} - {count} повторений")
```

```python
data = data[(data['Area'] < 500)]
data = data[(data['Kitchen area'] < 50)]
data = data[(data['Minutes to metro'] < 40)]
data = data[(data['Floor'] < 50)]
data = data[(data['Number of floors'] < 60)]
data = data[(data['Living area'] < 230)]
```
